```python
#4.1
secret = 4
guess = 6
if guess < secret:
    print("Too low!")
elif guess > secret:
    print("Too high!")
else:
    print("Just right!")
print("\n")

#4.2 cherry, pea, watermelon, pumpkin
small = False
green = True
if small and green:
    print("Pea")
elif small and not green:
    print("Cherry")
elif not small and green:
    print("Watermelon")
else:
    print("Pumpkin")
print("\n")

#6.1 3,2,1,0
for num in range(3,-1,-1):
    print(num)
print("\n")

#6.2
guess_me = 7
number =  1
while True:
    if number < guess_me:
        print("Too low!")
    elif number == guess_me:
        print("Found it!")
        break
    else:
        print("Oops")
        break
    number += 1
print("\n")

#6.3
guess_me = 5
for number in range(10):
    if number < guess_me:
        print ("Too low", number)
    elif number == guess_me:
        print("Found it!", number)
        break
    else:
        print("Oops")
        break
```
